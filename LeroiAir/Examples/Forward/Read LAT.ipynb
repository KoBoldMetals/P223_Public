More tests around how to read M?1 format data

```python
iname = "LAT_Tempest.mf1"
```

```python
#matlab inline
import os
import numpy
import pandas
import math
import matplotlib.pyplot as mpl
# import plotnine as p9
fs = (10, 6.18)
```

```python
os.stat(iname)
```

```python
f = open(iname, "r")
irawd = f.readlines()
f.close
print(len(irawd))
```

```python
icmnt = []
itemp = []
```

```python
for jl in irawd:
    if jl == "\n":
        continue
    else:
        temp = jl.split()
        if temp[0] == '/':
            icmnt.append(temp)
        elif temp[0].isnumeric():
            itemp.append(temp)
# convert the whole array at once
idata = numpy.array(itemp, dtype=float)
```

Create a data frame ...

```python
for jl in icmnt:
    if jl[1] == "DatIndx":
        ihedr = jl
        break
ihedr = ihedr[1:]
ifram = pandas.DataFrame(data=idata, index=idata[:,0], columns=ihedr)
```

Sort gates out

```python
# gates
for jl in icmnt:
    if jl[1] == "TIMES(ms)=":
        itime = jl
        break
itime = numpy.array(itime[2:], dtype=float)
NGATE = len(itime)
```

```python
# model header & values ...
for jl in icmnt: 
    if len(jl) >= 2: 
        if jl[1] == "SIG_T_1":
            mhedr = jl
        elif jl[1] == "MODEL_00":
            mvals = jl
        elif jl[1].startswith('UNITS'):
            munit = jl[1]
    else:
        continue

mhedr = mhedr[1:]
mvals = numpy.array(mvals[2:], dtype=float)
mfram = pandas.DataFrame(data=mvals, index=mhedr, columns=["Model 00"])
munit = munit.split("=")[1]
```

Now we need to figure out how to isolate inline & vertical data.
One option is to create two more data frames, indexed by icord

```python
txe = ifram['EastTx']
txn = ifram['NorthTx']
rxe = ifram['EastRx']
rxn = ifram['NorthRx']
rnge = txe.max()- txe.min()
rngn = txn.max()- txn.min()
if rnge > (1.1 * rngn):
    crd = txe
else:
    crd = txn
# print(crd)
```

Split data into components

```python
ib = "CHX"+str(1).zfill(3)
ie = "CHX"+str(NGATE).zfill(3)
vb = "CHZ"+str(1).zfill(3)
ve = "CHZ"+str(NGATE).zfill(3)
gx = []
gz = []
for jg in range(1, NGATE):
    gx.append("CHX"+str(jg).zfill(3))
    gz.append("CHZ"+str(jg).zfill(3))
```

We can put coordinates on the end because we reference the x-coordinate explicitly as well as the data set ...

```python
xfram = ifram.loc[:, ib:ie]
zfram = ifram.loc[:, vb:ve]
xfram = xfram.join(txe).join(rxe)
zfram = zfram.join(txe).join(rxe)
```

And because things are in a dataframe, plotting is relatively straightforward ...

```python
xplot = xfram.plot(x='EastTx', y=gx, title='Inline component', yticks=(.001, .01, .1, 1, 10,100), fontsize=12, logy=True, cmap='plasma', legend=False, grid=True, figsize=(10, 6.18));
xplot.figure.savefig('test_x.pdf')
```

```python
zplot = zfram.plot(x = 'EastRx', y = gz, title='Vertical component', yticks=(.001, .01, .1, 1, 10,100), fontsize=12, logy=True, cmap='plasma', legend=False, grid=True, figsize=fs);
zplot.figure.savefig('test_z.pdf')
```
